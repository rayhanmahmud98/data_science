```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = ("tips.csv")
df = pd.read_csv(file_path)
df
```

```python
sns.pairplot(df,hue="sex")
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","size"],hue="sex")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="kde")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="scatter")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="reg")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="hist")
plt.show()
```

```python
sns.pairplot(df,hue="sex",hue_order=["Female","Male"],kind="hist")
plt.show()
```
