```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
file_path = ("penguins.csv")
df = pd.read_csv(file_path).head(500)
df
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
plt.show()
```

```python
# order_1 =["Dream","Torgersen","Biscoe"]
# sns.barplot(x="island",y="sex",data=df,hue="sex",hue_order=["FEMALE","MALE"],order=order_1)
# plt.show()

order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",hue_order=["FEMALE","MALE"],order=order_1)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,orient="v")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,orient="h")
plt.show()
```
