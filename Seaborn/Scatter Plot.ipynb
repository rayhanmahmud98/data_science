```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = ("penguins.csv")
df = pd.read_csv(file_path)
df
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex",palette="magma",sizes=(100,40))
plt.show()
```
