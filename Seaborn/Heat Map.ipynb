```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = np.linspace(1,100,96).reshape(8,12)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
file_path = ("anagrams.csv")
df = pd.read_csv(file_path).head(30)
df
```

```python
dt = df.drop(columns=["attnr"],axis=1)
dt
```

```python
sns.heatmap(dt)
plt.show()
```

```python
sns.heatmap(dt,vmin=5,vmax=12)
plt.show()
```

```python
sns.heatmap(dt,vmin=1,vmax=12,cmap="tab10")
plt.show()
```

```python
sns.heatmap(dt,vmin=1,vmax=12,cmap="tab10",annot=True)
plt.show()
```

```python
var = np.linspace(1,15,15).reshape(3,5)
var
```

```python
sns.heatmap(var,vmin=1,vmax=12,cmap="tab10",annot=True)
plt.show()
```

```python
y = {"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=1,vmax=12,cmap="tab10",annot=True,annot_kws=y)
plt.show()
```

```python
y = {"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=1,vmax=12,cmap="tab10",annot=True,annot_kws=y,linewidth=5)
plt.show()
```

```python
sns.heatmap(var,vmin=1,vmax=12,cmap="tab10",annot=True,linewidth=5,cbar=False)
plt.show()
```

```python
x = sns.heatmap(var,vmin=1,vmax=12,cmap="tab10",annot=True,linewidth=5,cbar=False)
x.set(xlabel="python",ylabel="java")
plt.show()
```
